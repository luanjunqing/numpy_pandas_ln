```python
import pandas as pd
```

```python
# merging two df by keys.
left = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                      'A':['A0', 'A1', 'A2', 'A3'],
                      'B':['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                      'C':['C0', 'C1', 'C2', 'C3'],
                      'D':['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
res = pd.merge(left, right, on='key')
res
```

```python
left1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                             'key2': ['K0', 'K1', 'K0', 'K1'],
                             'A': ['A0', 'A1', 'A2', 'A3'],
                             'B': ['B0', 'B1', 'B2', 'B3']})
right1 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                              'key2': ['K0', 'K0', 'K0', 'K0'],
                              'C': ['C0', 'C1', 'C2', 'C3'],
                              'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left1
```

```python
right1
```

```python
# how = ['left', 'right', 'inner', 'outer']
res2 = pd.merge(left1, right1, on=['key1', 'key2'], how='left')
res2
```

```python
res3 = pd.merge(left1, right1, on=['key1', 'key2'], how='right')
res3
```
